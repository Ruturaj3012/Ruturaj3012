1.Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable. For example, if your categorical variable is age groups and
quantitative variable is income, then provide summary statistics of income grouped by the age groups. Create a list that contains a numeric value for each response to the categorical variable

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset from an open-source URL
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Display the first few rows of the dataset
print("Titanic Dataset:\n", df.head())
```

```python
# 1. Summary Statistics for Age grouped by 'Pclass' (as an example of a categorical variable)
summary_stats_age = df.groupby('Pclass')['Age'].describe()

# Display the summary statistics
print("\nSummary Statistics for Age Grouped by Pclass:\n", summary_stats_age)
```

```python
# 2. Summary Statistics for Fare grouped by 'Pclass'
summary_stats_fare = df.groupby('Pclass')['Fare'].describe()

# Display the summary statistics
print("\nSummary Statistics for Fare Grouped by Pclass:\n", summary_stats_fare)
```

```python
# Visualization 1: Distribution of Age for different Pclass
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Pclass', multiple="stack", kde=True)
plt.title('Age Distribution by Pclass')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Pclass')
plt.show()
```

```python
# Visualization 2: Boxplot of Age by Pclass
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Boxplot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.show()
```

```python
# Visualization 3: Distribution of Fare for different Pclass
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Fare', hue='Pclass', multiple="stack", kde=True)
plt.title('Fare Distribution by Pclass')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.legend(title='Pclass')
plt.show()
```

```python
# Visualization 4: Boxplot of Fare by Pclass
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Boxplot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

2. Write a Python program to display some basic statistical details like percentile, mean,
standard deviation etc. of the species of Iris-setosa, Iris-versicolor and Iris-versicolor
of iris.csv dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris dataset from an open-source URL
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv'
df = pd.read_csv(url, header=None)
df.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']

# Display the first few rows of the dataset
print("Iris Dataset:\n", df.head())
```

```python
# 1. Summary Statistics for numeric columns grouped by 'Species'
summary_stats = df.groupby('Species').agg(['mean', 'median', 'min', 'max', 'std'])

# Display the summary statistics
print("\nSummary Statistics Grouped by Species:\n", summary_stats)
```

```python
# 2. Percentiles for each species (25th, 50th, 75th percentiles)
percentiles = df.groupby('Species').quantile([0.25, 0.5, 0.75])

# Display the percentiles
print("\nPercentiles (25th, 50th, 75th) for each Species:\n", percentiles)
```

```python
# ---------------------------
# Visualizations:
# ---------------------------

# 1. Histogram of Sepal Length by Species
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='SepalLength', hue='Species', kde=True, multiple="stack")
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.legend(title='Species')
plt.show()
```

```python
# 2. Boxplot of Sepal Length by Species
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='SepalLength', data=df)
plt.title('Boxplot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# 3. Scatter plot between Petal Length and Petal Width by Species
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PetalLength', y='PetalWidth', hue='Species', style='Species', s=100)
plt.title('Scatter Plot of Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend(title='Species')
plt.show()
```

```python
# 4. Pairplot for the entire dataset
sns.pairplot(df, hue='Species')
plt.suptitle('Pairplot of Iris Dataset', size=16)
plt.show()
```

Descriptive Statistics:
Descriptive statistics summarize and organize the characteristics of a data set. The most commonly used descriptive statistics include:
Mean: The average value.
Median: The middle value when the data is sorted.
Minimum & Maximum: The smallest and largest values.
Standard Deviation (std): Measures the amount of variation or dispersion from the mean.
These statistics help understand the central tendency and spread of data.
 Grouped Statistics:
Grouped descriptive statistics allow us to compare how numeric variables behave across different categories. For example:
In the Titanic dataset, we analyzed the Age and Fare grouped by Pclass.
In the Iris dataset, we computed statistics for Sepal and Petal dimensions grouped by Species.
This grouping helps identify patterns and differences among categories.
 Visualization Techniques:
Visualizations enhance understanding of statistical results:
Histograms: Show the distribution of a variable and its frequency.
Boxplots: Display median, quartiles, and outliers to show data spread and skewness.
Scatter plots: Help to examine relationships between two numeric variables.
Pairplots: Show scatter plots for all variable pairs and histograms on the diagonal.
 Observations:
Titanic Dataset:
The mean age varied across passenger classes, with 1st class passengers generally older.
Fare distribution showed higher variance in 1st class compared to 3rd class.
Visuals clearly showed age and fare differences among classes.
Iris Dataset:
The Setosa species had smaller petal lengths and widths compared to Versicolor and Virginica.
Boxplots showed clear separation among species.
Pairplots and scatter plots revealed linear relationships between petal length and width.
 Conclusion:
Descriptive statistics and visualizations provide critical insights into dataset characteristics. Grouped summaries help compare numeric variables across categories, while visual tools help identify trends, outliers, and patterns in the data.
